## Time Series
### Using Moving Averages

```python
# create an array for each product to make predictions for
p = []
for col in df_10.columns:
    p.append(df_10.loc[:,col])
```

```python
def plot_series(time, series, format="-", start=0, end=None):
    ''' 
    this function is used to plot the time series
    INPUT:
    time - the number of weeks in an array
    series - the values being plot
    format - default line plot
    OUTPUT: a plot of the time series
    '''
    plt.plot(time[start:end], series[start:end], format, label='Predicted')
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(True)
def moving_average_predict(series, window_size):
    '''
    this function gets our predictions using a moving average approach
    INPUTS:
    series - the predictions in an array
    window_size - the number of weeks in the window frame being used to calculate the moving average
    OUTPUT: an array of the predicted values
    '''
    predictions = []
    for time in range(len(series) - window_size):
        predictions.append(series[time:time + window_size].mean())
    return np.array(predictions)
```

### Visualization

```python
split_time = 40 #train on the first 40 weeks
# plot the time series of the actual and the predicted for each product
for i in range(len(p)):
    moving_avg = moving_average_predict(p[i], 5)[split_time - 5:]
    plt.figure(figsize=(9, 5))
    plt.title('Product' + str(i+1))
    p[i][split_time:].plot(label='Actual')
    plot_series(p[i][split_time:].index, moving_avg)
    plt.legend()
```

### Evaluation
In this analysis we will use the mean absolute error. The mean absolute error (MAE) is the simplest regression error metric to understand. Weâ€™ll calculate the residual for every data point, taking only the absolute value of each so that negative and positive residuals do not cancel out. We then take the average of all these residuals. Effectively, MAE describes the typical magnitude of the residuals. The MAE is also the most intuitive of the metrics since weâ€™re just looking at the absolute difference between the data and the modelâ€™s predictions

```python
mae = [] # create an array to hold the mae for each product
for i in range(len(p)):
    moving_avg = moving_average_predict(p[i], 5)[split_time - 5:]
    mae.append(mean_absolute_error(p[i][split_time:], moving_avg))
```

```python
print("The average mean absolute error for the first 10 products is", sum(mae)/len(mae))
```

## ARIMA

It is important to do diagnostics such as check the components of the time series by looking at the autocorrelation plots, seasonality, trend and residuals. However, it would take a long time to do that for each product so instead I experimented with the p,d,q parameters and chose the best one.

```python
pred = [] # create an array to hold the predictions for each product
for i in range(len(p)):
    train, test = p[i][:split_time], p[i][split_time:] # train on the first 40 weeks and test on the last 12
    history = [x for x in train] # store the actuals
    predictions = [] # obtain the predictions using arima
    for t in range(len(test)):
        model = ARIMA(history, order=(5,1,0)) # can only experiment with p,d values since changing q will make the model not converge
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        prediction = output[0]
        predictions.append(prediction)
        pred.append(predictions)
        actual = test[t]
        history.append(actual)
```

### Visualization

```python
for i in range(len(p)):
    plt.figure(figsize=(9, 5))
    plt.title('Product' + str(i+1))
    p[i][split_time:].plot()
    plot_series(p[i][split_time:].index, pred[i])
```

## Evaluation

## Using FBProphet

```python
# first we need to adjust the formatting before being able to use fbprophet
# we need to make sure the index is a datetime object
# disregard the year in the datetime we are only concerned with the week
# we also need to rename the columns to 'ds' and 'y' to comply with fbprophet
p_fb = []
for i in range(len(p)):
    p[i].index = pd.to_datetime(list(range(0,52)), unit='W')
    p_fb.append(p[i].reset_index().rename(columns = {'index': 'ds',p[i].reset_index().columns[1]: 'y'}))
```

```python
# create and fit the model to each product
models = []
for i in range(len(p_fb)):
    model = Prophet(weekly_seasonality=True) # we only want the weekly seasonality since it is weekly data
    models.append(model.fit(p_fb[i]))
```

```python
# make forecasts for each product
forecasts = []
for model in models:
    forecast = model.make_future_dataframe(periods=8, freq='W') # we want the next 7 values (by week)
    forecasts.append(model.predict(forecast))
```

```python
forecasts[0][['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head() # example of the forecasted values for the first product
```

```python
models[0].plot(forecasts[0], xlabel = 'Date', ylabel = 'Volume') # plotting tbe model and the forecasted values
```

```python
from fbprophet.diagnostics import cross_validation
df_cv = []
for model in models:
    df_cv.append(cross_validation(model, initial='280 days', period='7 days', horizon = '7 days')) # use the first 280 days
    #to train on and predict the remaining weeks
df_cv[0].head()
```

```python
# obtain the performance metrics for each model
from fbprophet.diagnostics import performance_metrics
df_p = []
for cv in df_cv:
    df_p.append(performance_metrics(cv))
```

```python
mae = []
for p in df_p:
    mae.append(p['mae'].values)
```

```python
print("The average mean absolute error for the first 10 products is",(sum(mae)/len(mae))[0])
```

## Using Machine Learning

```python
# first melt the dataset in order to get a format where we can do regression on
# this involves melting the dataset
melt = df1.iloc[0:10,:].melt(id_vars='Product_Code', var_name='Week', value_name='Sales')

melt['Product_Code'] = melt['Product_Code'].str.extract('(\d+)', expand=False).astype(int) # make things easier by removing letters
melt['Week'] = melt['Week'].str.extract('(\d+)', expand=False).astype(int) # remove letters

melt = melt.sort_values(['Week', 'Product_Code']) # get the melted dataset of the top 10
melt.head()
```

```python
# now use the dataset above to create new variables
melt2 = melt.copy()
melt2['Last_Week_Sales'] = melt2.groupby(['Product_Code'])['Sales'].shift()
melt2['Last_Week_Diff'] = melt2.groupby(['Product_Code'])['Last_Week_Sales'].diff() # create variable for weekly differencing
melt2 = melt2.dropna()
melt2.head()
```

```python
mae = []
for week in range(40,52):
    train = melt2[melt2['Week'] < week] # create train set by using previous weeks to predict the next
    val = melt2[melt2['Week'] == week] # create validation set (current week)

    xtr, xts = train.drop(['Sales'], axis=1), val.drop(['Sales'], axis=1) # drop the sales feature and create x train and x test
    ytr, yts = train['Sales'].values, val['Sales'].values # only keep the sales as our response var for y train and y test

    m = RandomForestRegressor(n_estimators=1000, n_jobs=-1, random_state=0) # use a random forest regression model
    m.fit(xtr, ytr) # fit the model

    p = m.predict(xts) # predict on the current week

    error = mean_absolute_error(yts, p) # obtain the mae
    print('Week %d - Error %.5f' % (week, error))
    mae.append(error)
print('Mean Absolute Error = %.5f' % np.mean(mae))
```

We could make the model better by using cross validation to hyptertune the parameters or by using a log transform or adding more features such as lags. We could also try another algorithm such as xgboost. For now we will just try using the log transform.

```python
# taking a log transform
mae = []
for week in range(40,52):
    train = melt2[melt2['Week'] < week]
    val = melt2[melt2['Week'] == week]

    xtr, xts = train.drop(['Sales'], axis=1), val.drop(['Sales'], axis=1)
    ytr, yts = train['Sales'].values, val['Sales'].values

    m = RandomForestRegressor(n_estimators=1000, n_jobs=-1, random_state=0)
    m.fit(xtr, np.log1p(ytr))

    p = np.expm1(m.predict(xts))

    error = mean_absolute_error(yts, p)
    print('Week %d - Error %.5f' % (week, error))
    mae.append(error)
print('Mean Error = %.5f' % np.mean(mae))
```

The model did improve and we can probably make it better but instead we will compare the results of the different techniques
